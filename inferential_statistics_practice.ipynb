## **Binomial Distribution**

### **Q1. A LED bulb manufacturing company regularly conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for LED light bulbs, that the company manufactures is 5%. Suppose a random sample of 10 LED light bulbs is selected. Find the probability distribution for the defective bulbs and answer the following questions.**

**a) What is the probability that none of the LED bulbs are defective?**

```python
# Q1a: Probability that none of the LED bulbs are defective
# n = 10, p = 0.05 (failure rate), k = 0
prob_none_defective = stats.binom.pmf(0, n=10, p=0.05)
print(f"Probability that none of the LED bulbs are defective: {prob_none_defective:.4f}")
```

**Observation:**
- The probability that none of the LED bulbs are defective is approximately **59.87%**.
- This is relatively high, which makes sense given the low failure rate (5%) and small sample size (10 bulbs).
- This probability represents the likelihood of getting 0 defective bulbs out of 10 when the failure rate is 5%.


**b) What is the probability that exactly one LED bulb is defective?**

```python
# Q1b: Probability that exactly one LED bulb is defective
# n = 10, p = 0.05, k = 1
prob_one_defective = stats.binom.pmf(1, n=10, p=0.05)
print(f"Probability that exactly one LED bulb is defective: {prob_one_defective:.4f}")
```

**c) What is the probability that two or fewer LED bulbs are defective?**

**Hint:** We need to calculate the cumulative probability of two or fewer LED bulbs being defective.

```python
# Q1c: Probability that two or fewer LED bulbs are defective
# n = 10, p = 0.05, k <= 2
prob_two_or_fewer = stats.binom.cdf(2, n=10, p=0.05)
print(f"Probability that two or fewer LED bulbs are defective: {prob_two_or_fewer:.4f}")
```

**d) What is the probability that three or more LED bulbs are defective?**

**Hint:** We need to subtract the cumulative probability of up to 2 defective LED bulbs from 1.

```python
# Q1d: Probability that three or more LED bulbs are defective
# P(X >= 3) = 1 - P(X <= 2)
prob_three_or_more = 1 - stats.binom.cdf(2, n=10, p=0.05)
print(f"Probability that three or more LED bulbs are defective: {prob_three_or_more:.4f}")
```

## **Uniform Distribution**

### **Q4. A university records the time taken by students to solve the statistics assignment in 'assignment.csvâ€™. Plot the probability distribution this data follows and answer the following questions.**

### **Loading the dataset**

```python
assignment = pd.read_csv('assignment.csv')

assignment.head()

# Plot the probability distribution (Uniform Distribution)
# Determine the parameters
a = assignment['Time_taken'].min()
b = assignment['Time_taken'].max()

# Create a range of values for plotting
x = np.linspace(a - 0.5, b + 0.5, 1000)
# Uniform PDF: f(x) = 1/(b-a) for a <= x <= b
pdf_values = stats.uniform.pdf(x, loc=a, scale=b-a)

# Plot the uniform distribution
fig = plt.figure(figsize=(10, 6), dpi=300)
plt.plot(x, pdf_values, 'b-', linewidth=2, label='Uniform PDF')
# Removed shading; do not use fill_between
plt.xlabel('Time taken (hours)', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.title('Uniform Distribution of Time Taken to Complete Assignment', fontsize=14)
plt.axvline(a, color='r', linestyle='--', label=f'Min = {a:.2f}')
plt.axvline(b, color='r', linestyle='--', label=f'Max = {b:.2f}')
plt.legend()
# Removed gridlines; do not use plt.grid
plt.show()

print(f"Uniform Distribution Parameters:")
print(f"Minimum (a): {a:.2f} hours")
print(f"Maximum (b): {b:.2f} hours")
print(f"Range: {b-a:.2f} hours")
```

**a) Find the probability that a randomly selected student requires at most 2.5 hours to complete the assignment.**

**CDF** of a random variable (X) is the probability that X will take the value less than or equal to x. Mathematically, it can be represented as:

>$F_X(x) = P(X\leq x)$

In our case, the random variable (X) is the number of hours.

$ P(X\leq 2.5)$

```python
# Q4a: Probability that a randomly selected student requires at most 2.5 hours
# First, we need to determine the parameters of the uniform distribution
# For uniform distribution: a = min(Time_taken), b = max(Time_taken)
a = assignment['Time_taken'].min()
b = assignment['Time_taken'].max()
print(f"Uniform distribution parameters: a = {a:.2f}, b = {b:.2f}")

# P(X <= 2.5) using uniform CDF
prob_at_most_2_5 = stats.uniform.cdf(2.5, loc=a, scale=b-a)
print(f"\nProbability that a student requires at most 2.5 hours: {prob_at_most_2_5:.4f}")
```

**b) Find the probability that a randomly selected student requires at least 3 hours to complete the quiz.**

$ P(X >= 3)$

```python
# Q4b: Probability that a randomly selected student requires at least 3 hours
# P(X >= 3) = 1 - P(X < 3)
prob_at_least_3 = 1 - stats.uniform.cdf(3, loc=a, scale=b-a)
print(f"Probability that a student requires at least 3 hours: {prob_at_least_3:.4f}")
```

**Observation:**
- The probability that a student requires at least 3 hours is approximately **33.33%** (0.3333).
- This represents the upper third of the distribution (from 3 to 4 hours out of the total 3-hour range).
- Since the distribution is uniform, the probability is simply the proportion of the range: (4 - 3) / (4 - 1) = 1/3 â‰ˆ 0.3333.
- This means about one-third of students take 3 hours or more to complete the assignment, which could be useful for planning assignment deadlines or identifying students who might need additional support.


**c) Find the probability that a randomly selected student requires 1.5 to 3.5 hours to complete the quiz.**

$ P(1.5 <= X <= 3.5)$

```python
# Q4c: Probability that a randomly selected student requires 1.5 to 3.5 hours
# P(1.5 <= X <= 3.5) = P(X <= 3.5) - P(X < 1.5)
prob_between = stats.uniform.cdf(3.5, loc=a, scale=b-a) - stats.uniform.cdf(1.5, loc=a, scale=b-a)
print(f"Probability that a student requires 1.5 to 3.5 hours: {prob_between:.4f}")
```

### **Q8. Suppose an automobile battery manufacturer claims that the mean lifetime of their battery is 60 months with a standard deviation of 6 months. Suppose the distribution of battery life is approximately normal. Find the probability that the mean lifetime of 40 randomly sampled batteries will be less than 58 months.**

```python
# Q8: Probability that the mean lifetime of 40 randomly sampled batteries will be less than 58 months
# Population: mean = 60, std = 6
# Sample size n = 40
# Sampling distribution of sample mean: mean = 60, std_error = 6/sqrt(40)
pop_mean = 60
pop_std = 6
n = 40
std_error = pop_std / np.sqrt(n)

prob_mean_less_58 = stats.norm.cdf(58, loc=pop_mean, scale=std_error)
print(f"Probability that the mean lifetime of 40 randomly sampled batteries will be less than 58 months: {prob_mean_less_58:.4f}")
```

### **Q9. A random sample of 40 households was selected as part of a study on electricity usage, and the number of kilowatt-hours (kWh) was recorded for each household in the sample for the first quarter of 2022. The average usage was found to be 310 kWh. In a very large study in the first quarter of the previous year, it was found that the standard deviation of the usage was 89 kWh.**

### **Assuming the standard deviation is unchanged and that the usage is normally distributed, provide an expression for calculating the 95% confidence interval for the mean usage in the first quarter of 2022.**

```python
# Q9: 95% confidence interval for the mean usage
# Sample: n = 40, x_bar = 310
# Population: std = 89 (known)
# For 95% CI: z_alpha/2 = 1.96
n_sample = 40
x_bar = 310
pop_std_known = 89
z_critical = stats.norm.ppf(0.975)  # 95% confidence level, two-tailed

# Confidence interval formula: x_bar Â± z_alpha/2 * (std / sqrt(n))
margin_of_error = z_critical * (pop_std_known / np.sqrt(n_sample))
lower_bound = x_bar - margin_of_error
upper_bound = x_bar + margin_of_error

print(f"95% Confidence Interval for the mean usage:")
print(f"Lower bound: {lower_bound:.2f} kWh")
print(f"Upper bound: {upper_bound:.2f} kWh")
print(f"\nExpression: {x_bar} Â± {z_critical:.2f} Ã— ({pop_std_known} / âˆš{n_sample})")
print(f"Or: {x_bar} Â± {margin_of_error:.2f}")
```

## **Hypothesis Testing**

**Observation:**
- The 95% confidence interval for mean electricity usage is **[282.42, 337.58] kWh**.
- This means we are 95% confident that the true population mean lies within this range.
- The margin of error is **Â±27.58 kWh**, calculated using the z-critical value (1.96) and standard error (89/âˆš40 â‰ˆ 14.07).
- The interval is centered at the sample mean (310 kWh), which is our best estimate of the population mean.
- The width of the interval (55.16 kWh) reflects the uncertainty in our estimate, influenced by sample size (n=40) and population standard deviation (89 kWh).
- This confidence interval is useful for energy planning, budgeting, and understanding typical household electricity consumption patterns.

### **Q10. You are the manager of a Chinese restaurant. You want to determine whether the waiting time to place an order has changed in the past month from its previous population mean value of 4.5 minutes. State the null and alternative hypothesis.**

```python
# Q10: State the null and alternative hypothesis
# H0: The waiting time has not changed from 4.5 minutes (Î¼ = 4.5)
# H1: The waiting time has changed from 4.5 minutes (Î¼ â‰  4.5)

print("Null Hypothesis (H0): Î¼ = 4.5 minutes")
print("Alternative Hypothesis (H1): Î¼ â‰  4.5 minutes")
print("\nThis is a two-tailed test since we want to determine if the waiting time has changed (either increased or decreased).")
```

**Key Insights:**
- **Null Hypothesis (H0): Î¼ = 4.5 minutes** - This represents the status quo or the claim that waiting time has not changed.
- **Alternative Hypothesis (H1): Î¼ â‰  4.5 minutes** - This represents the research question that waiting time has changed.
- This is a **two-tailed test** because we're testing for any change (either increase or decrease), not just one direction.
- The two-tailed approach is appropriate when we want to detect any significant change from the historical mean, regardless of direction.
- This hypothesis setup allows the manager to determine if recent changes (staffing, menu, processes) have affected waiting times in either direction.
- The test will help make data-driven decisions about whether operational changes are needed.


**Observation:**
- The p-value for a two-tailed test with Z-stat = +2.00 is approximately **0.0455** (4.55%).
- This p-value represents the probability of observing a test statistic as extreme as Â±2.00 or more extreme, assuming the null hypothesis is true.
- At a common significance level of Î± = 0.05, we would **reject the null hypothesis** since p-value (0.0455) < Î± (0.05).
- A Z-statistic of 2.00 means the sample mean is 2 standard errors away from the hypothesized population mean.
- This result suggests there is statistically significant evidence that the population parameter differs from the null hypothesis value.
- The p-value of 0.0455 indicates moderate evidence against the null hypothesis - not extremely strong, but sufficient to reject at the 5% level.
- In practical terms, this p-value suggests that if the null hypothesis were true, we would observe such an extreme result only about 4.55% of the time by chance alone.


### **Q11. Find the p-value of a two-tailed hypothesis test if the Z-stat = +2.00?**

```python
# Q11: Find the p-value of a two-tailed hypothesis test if Z-stat = +2.00
z_stat = 2.00

# For a two-tailed test, p-value = 2 * P(Z > |z_stat|)
# Since z_stat is positive, P(Z > 2.00) = 1 - P(Z <= 2.00)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
print(f"P-value for a two-tailed test with Z-stat = {z_stat}: {p_value:.4f}")
print(f"\nInterpretation: The probability of observing a test statistic as extreme as {z_stat} or more extreme, assuming the null hypothesis is true.")
```
